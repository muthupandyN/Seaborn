# seaborn


```python
import seaborn as sns
df=sns.load_dataset("D:tips")
df.head()
```

# correalation with heatmap
The correlation heat map uses color cells,it typycally in monochromatic scale in 2d corelation matrix,which is used in feature selection

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# univariate analysis

```python
sns.jointplot(x='size',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# pairplot
pairplot is also known as scatterplot,one row of data is matched with another variable values

```python
sns.pairplot(df)
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```



# Distplot
distplot help us to check distribution of the column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```
